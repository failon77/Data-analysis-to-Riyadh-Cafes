```python
import pandas as pd
```

```python
df = pd.read_csv('data/riyadh_cafes.csv')
df.head()
```

```python
#حذف العمود الاول
df=df.drop("index", axis=1)
```

```python
df.info()
```

```python
df.dtypes
```

```python
#الاستعلام عن القيم المفقودة
print(df.isnull().sum())
```

```python
# التحقق من وجود بعض القيم الفارغة المكتوبه بصيغة string
df_nul= df[df['lan']=="Null"]
df_nul.head()
```

```python
#وضع القيم الصحيحة في متغير df_not_nul و تصفية القيم المفقودة
df_not_nul=df[df['lan']!="Null"]
df_not_nul.head()
```

```python
#تحويل القيم الموجودة في العموديين الاخيرة الى قيم عشرية لتمكين معالجتها
df_not_nul["lan"]=df_not_nul["lan"].astype(float)
df_not_nul["lon"]=df_not_nul["lon"].astype(float)
```

```python
# حذف عمود رابط الموقع
df_not_nul= df_not_nul.drop('url',axis=1)
df_not_nul.head()
```

```python
# تصفية المقاهي التي تحتوي على اقل من 50 مراجع
df_rate=df_not_nul[(df_not_nul['rating_count']>50)] 
df_rate.head()
```

```python
#تحويل القيم الموجودة من نصية الى عددية في عمود (24_hours)

df_rate.loc[df_rate['24_hours']==True,'24_hours']=1
df_rate.loc[df_rate['24_hours']==False,'24_hours']=0
df_rate.head()
```

```python
#الاستعلام عن اسم اقل تقييم كافيه
df_rate[df_rate['rating'] == df_rate['rating'].min()]
```

```python
#الاستعلام عن اسم اعلى تقييم كافيه
df_rate[df_rate['rating'] == df_rate['rating'].max()]
```

```python
#الاستعلام عن اعلى عدد تقييم لكافيه
df_rate[df_rate['rating_count'] == df_rate['rating_count'].max()]
```

```python
#القيم المتفردة
df_rate['coffeeName'].nunique()
```

```python
#الاستعلام عن عدد الكافيهات المفتوحة 24 ساعة 
df_rate['24_hours'].value_counts().head()
```

```python
#معرفة متوسط عدد التقييمات لكل من الكافيهات المفتوحة 24 ساعة و كافيهات محدودة المدة
df_rate.groupby('24_hours').mean()['rating_count']
```

```python
import folium
from IPython.display import display

# عرض مواقع الكافيهات على الخريطة
def cart(df):

    map_osm = folium.Map(location=[df["lan"].mean(),df["lon"].mean()], zoom_start=10)
    for index, row in df.iterrows():
        marker = folium.Marker(
            location=[row['lan'],row['lon']],
            popup=row['coffeeName']
        )
        marker.add_to(map_osm)

    display(map_osm)
cart(df_rate)
```

```python
#تقسيم البيانات لمجموعتين 

DF_N1=df_rate[df_rate['24_hours']!= 1] #مجموعة الكافيهات التي تفتح لمدة محددة 
DF_N0=df_rate[df_rate['24_hours']!= 0] #مجموعة الكافيهات التي تفتح ل24 ساعة
```

```python


#عرض مواقع الكافيهات التي تفتح لمدة محددة على الخريطة
def cart(df):
    map_osm = folium.Map(location=[df["lan"].mean(),df["lon"].mean()], zoom_start=10)
    for index, row in df.iterrows():
        marker = folium.Circle(color="green",
            location=[row['lan'],row['lon']],
           
            popup=row['coffeeName']
        )
        marker.add_to(map_osm)

    display(map_osm)
cart(DF_N1)
```
